### Por Edad

Para este caso se presentan valores NAN por lo que se tiene que tomar una decision si se eliminan o reemplazan por un valor. Se puede observar que la caracteristica 'Name' lleva un abreviacion como Mr., Miss., Mrs, entre otros por lo que se puede generar una relacion entre dicho string con la edad.

```python
group = df[['Pclass', 'Survived','PassengerId', 'Sex', 'Name', 'Age']]
group.sample(10)
```

```python
for i in range(group.shape[0]):
    c = group.Name[i].split(',')[1]
    group.Name[i] = c.split('.')[0]
```

```python
group.sample(10)
```

```python
group1 = group[['Name', 'Age']].groupby(['Name']).min()
group2 = group[['Name', 'Age']].groupby(['Name']).max()
```

```python
group.Age = group.Age.fillna(-1)
```

```python
for i in range(group.shape[0]):
    if group.Age[i] == -1:
        for j in range(group2.shape[0]):
            if group.Name[i] == group1.index[j]:
                group.Age[i] = int((group1.Age[j] + group2.Age[j])/2)
```

```python
group.sample(10)
```

```python
condition = (group['Survived'] == 1) & (group['Sex'] == 'female') & (group['Age'] > 17)
print("Sobrevivieron: ", group.loc[condition].shape[0], "mujeres mayores a 17 a√±os")
```

```python
condition = (group['Survived'] == 1) & (group['Sex'] == 'male') & (group['Age'] > 17)
print("Sobrevivieron: ", group.loc[condition].shape[0], "varones mayores a 17 a√±os")
```

```python
condition = (group['Survived'] == 1) & (group['Age'] < 17)
print("Sobrevivieron: ", group.loc[condition].shape[0], "personas menores a 17 a√±os")
```

## Analizando por parientes 
En este caso influye mucho si estas con parientes en el barco ya que tu vida depende de otras personas, por lo que se generara una nueva caracteristica parientes como un binario, si tienes o no parientes a bordo.

```python
parents = df[['SibSp', 'Parch']]
```

```python
parents['parentss'] = parents[['SibSp', 'Parch']].sum(axis=1) 
```

```python
parents['parentss'] = parents['parentss'].apply(lambda x : 0 if x==0 else 1)
```

```python
parents.head(10)
```

##### Concatenando con las caracteristicas cambiadas 

```python
df = df.drop(['Name', 'Age','SibSp','Parch'], axis=1)
```

```python
df = pd.concat([df, group[['Name','Age']], parents[['parentss']]], axis=1, ignore_index=False)
```

```python
df.head()
```

```python
df[['Survived','PassengerId', 'parentss']].groupby(['Survived','parentss']).count()
```

Se observa que hay mayor personas que no sobrevivieron que no tenian parientes a bordo.

```python
df[['Pclass','Survived','PassengerId', 'parentss']].groupby(['Pclass','parentss']).count()
```

Se observa que la clase 3 fue la que mayor numero de personas fue sin parientes a bordo, por lo que da mayor afirmacion a lo planteado anterior que son los mas propensos a no sobrevivir.
